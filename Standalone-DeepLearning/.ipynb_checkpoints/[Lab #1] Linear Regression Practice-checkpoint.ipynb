## 1. Data Generation (Preparation)

```python
X = list(range(10))
Y = [1, 1, 2, 4, 5, 7, 8, 9, 9, 10]
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, Y)
plt.plot(X, Y)
```

```python
class H():
    def __init__(self, w):
        self.w = w
        
    def forward(self, x):
        return self.w * x
    
        
def cost(h, x, y):
    error = 0
    for i in range(len(x)):
        error += (h.forward(x[i]) - y[i]) ** 2
    error /= len(x)
    return error
```

```python
h = H(4)
cost(h, X, Y) 
```

```python
list_w = [w * 0.1 for w in range(-100, 100)]
list_c = [cost(H(w), X, Y) for w in list_w]

plt.xlabel('w')
plt.ylabel('cost')
plt.plot(list_w, list_c)
```

```python
def cal_grad(w, cost):
    eps = 0.0001
    h = H(w)
    cost1 = cost(h, X, Y)
    h = H(w + eps)
    cost2 = cost(h, X, Y)
    dcost = cost2 - cost1
    dw = eps
    grad = dcost / dw
    return grad, (cost1+cost2)*0.5


def cal_grad2(w, cost):
    h = H(w)
    grad = 0
    for i in range(len(X)):
        grad += 2 * (h.forward(X[i]) - Y[i]) * X[i]
    grad = grad / len(X)
    c = cost(h, X, Y)
    return grad, c


lr = 0.001
w1 = 4
w2 = 4

list_w1 = []
list_c1 = []
list_w2 = []
list_c2 = []

for i in range(100):
    grad, mean_cost = cal_grad(w1, cost)
    grad2, mean_cost2 = cal_grad2(w2, cost)
    w1 -= lr * grad
    w2 -= lr * grad
    list_w1.append(w1)
    list_w2.append(w2)
    list_c1.append(mean_cost)
    list_c2.append(mean_cost2)
    print('{:.12f} {:.12f} {:.12f} {:.12f}'.format(w1, mean_cost, w2, mean_cost2))
```

```python
plt.scatter(list_w1, list_c1)
plt.scatter(list_w2, list_c2)
```
