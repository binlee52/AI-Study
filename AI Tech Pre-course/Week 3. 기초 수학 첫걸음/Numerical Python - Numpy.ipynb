### Array creation

파이썬은 Dynamic Typing으로 실행시점에 data type을 결정.
즉, 하나의 리스트에 다양한 데이터 타입이 들어갈 수 있다.
그러나 Numpy는 Dynamic Typing을 지원하지 않아서 속도가 빠르며,
data type 을 하나로 통일 해야한다.

```python
arr = np.array(["1", "4", 5, 8], float)

# numpy.float64메모리 공간의 크기를 64 bit 로 설정
arr, type(arr[0])
```

```python
# 모양, 데이터 타입
arr.shape, arr.dtype
```

```python
# 차원 개수, 원소의 총 개수
arr.ndim, arr.size
```

### Numpy 배열 생성

- arange
- ones
- zeros
- empty
- ones_like (shape 복사)
- identity (단위행렬)
- eyes (대각선이 1인 행렬)
- random sampling

```python
np.arange(10).reshape(-1, 5)
```

```python
np.arange(0, 5, 0.5)
```

```python
np.zeros(shape=(10, ), dtype=np.int8)
```

```python
np.ones(shape=(10, ), dtype=np.int8)
```

```python
np.empty(shape=(10, ), dtype=np.int8)
```

```python
arr= np.arange(30).reshape(-1, 6)
np.ones_like(arr)
```

```python
np.identity(n=3, dtype=np.int8)
```

```python
np.eye(N=3, M=5, dtype=np.int8)     # k -> start index
```

```python
# 대각 행렬의 값을 추출
arr = np.arange(9).reshape(3, 3)
np.diag(arr)    # k -> start index
```

```python
# 균등분포
np.random.uniform(0, 1, 10).reshape(2, 5)
```

```python
# 정규분포
np.random.normal(0, 1, 10).reshape(2, 5)
```

```python
# 합
arr, arr.sum()
```
